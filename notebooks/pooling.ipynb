# Pooling

El pooling es otra operación frecuente en las redes convolucionales, la cual está pensada para reducir el número de parámetrosc, reducir el costo computacional y encontrar características más robustas. 

```python
import os
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from IPython import display
```

```python
tf.enable_eager_execution()
```

En este cuaderno también usaremos las imágenes del edicio Coltejer y una obra de Fernando Botero

```python
images_path = os.path.join('..', 'images', 'N01_convoluciones')
botero_fpath = os.path.join(images_path, 'botero.jpg')
coltejer_fpath = os.path.join(images_path, 'coltejer.jpg')
```

```python
display.Image(coltejer_fpath, width=128)
```

```python
display.Image(botero_fpath, width=180)
```

```python
botero_image = plt.imread(botero_fpath).astype(np.float32)
coltejer_image = plt.imread(coltejer_fpath).astype(np.float32)
msg = "Dimensión de la imagen del edificio: {}"
print(msg.format(coltejer_image.shape))
msg = "Dimensión de la imagen del Botero: {}"
print(msg.format(botero_image.shape))
```

Primero veamos lo que el *pooling* le hace a una imagen (para lo que utilizaremos [tf.nn.pool](https://www.tensorflow.org/api_docs/python/tf/nn/pool)) y después la explicación

```python
botero_pooled = tf.nn.avg_pool(
    botero_image.reshape((1, *botero_image.shape)),
    [1, 8, 8, 1],
    [1, 1, 1, 1],
    padding='SAME'
).numpy()

coltejer_pooled = tf.nn.avg_pool(
    coltejer_image.reshape((1, *coltejer_image.shape)),
    [1, 8, 8, 1],
    [1, 1, 1, 1],
    padding='SAME'
).numpy()
```

```python
plt.imshow(botero_pooled[0].astype(np.uint8))
plt.axis('off')
```

```python
plt.imshow(coltejer_pooled[0].astype(np.uint8))
plt.axis('off')
```

Podemos ver que el efecto del *pooling* es volver un poco borrosa la imagen. Ahora veamos cómo funciona.

![imagen que explica max pooling](../images/N02_pooling/max_pooling.gif)

Para ser más precisos, la operación de la imágen se conoce como *max pooling*, mientras que nosotors utilizamos *average pooling*. El pooling consiste el tomar una vencidad de pixeles y resumirlos en uno solo, sea utilizando el promedio de los valores y el máximo de ellos. Independiente de cuál opción se escoja, el efecto es el mismo.
